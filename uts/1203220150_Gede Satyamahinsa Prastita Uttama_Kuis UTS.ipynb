4. Buatlah pemodelan dan simulasi sangat sederhana dari kejadian sekitar anda dengan memanfaatkan library ModSim menggunakan semua fungsi yang ada dalam library tersebut dan juga dalam bentuk excel untuk simulasi tersebut seperti simulasi monte carlo!

```python
import matplotlib.pyplot as plt
from modsim import *

def update_func(pop, system):
    # Fungsi update untuk menghitung populasi bakteri
    net_growth = system.growth_rate * pop
    return pop + net_growth

def run_simulation(system, update_func):
    # Jalankan simulasi pertumbuhan populasi bakteri
    results = TimeSeries()
    results[system.t_0] = system.p_0

    for t in linrange(system.t_0, system.t_end):
        results[t+1] = update_func(results[t], system)
    
    return results

# Menetapkan 3 parameter pada sistem
growth_rate = 0.1  # Laju pertumbuhan per satuan waktu
p_0 = 10           # Populasi awal
t_end = 48         # Durasi simulasi dalam satuan waktu (jam)

system = System(growth_rate=growth_rate, p_0=p_0, t_0=0, t_end=t_end)
results = run_simulation(system, update_func)
print(results)

# Melakukan plot terhadap results
plt.figure(figsize=(10, 5))
plt.plot(results.index, results.values, label='Model Pertumbuhan Bakteri')
decorate(xlabel='Waktu (jam)',
        ylabel='Populasi Bakteri',
        title='Simulasi Pertumbuhan Populasi Bakteri')
plt.legend()
plt.grid(True)
plt.show()
```
